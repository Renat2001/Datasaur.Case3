
## Prediction on Test Data

```python
def pred(file_index):
  res, = model(f'/content/drive/MyDrive/case3-datasaur-photo/techosmotr/techosmotr/test/{file_index}.jpeg')
  return res.probs.top1
```

```python
df['class'] = df['file_index'].apply(pred)
```
